```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.integrate import odeint
```

```python
df = pd.read_excel("Classeur1v2.xlsx")  
```

```python
df.head()
```

```python
t = np.array(df['t(s)'])
x = np.array(df['x(m)'])
```

```python
plt.plot(t,x, 'bo')
params = linregress(t,x)
plt.plot(t,params.slope*t+params.intercept, 'b')
plt.xlabel('temps t(s)')
plt.ylabel('abscisse x(m)')
plt.title('La trajectoire x(t) avec les valeurs experiementales')
```

```python
v = np.ones_like(t) *params.slope
plt.plot(t,v, 'g')
plt.xlabel('temps t(s)')
plt.ylabel('vitesse v(m/s)')
plt.title('La vitesse V(t) d\' apres les valeurs experiementales')
```
