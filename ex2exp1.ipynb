```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.integrate import odeint
```

```python
df = pd.read_excel('Classeur3.xlsx')
df.head()
```

```python
t = np.array(df['t(s)'])-1.90
x = np.array(df['x(m)'])
```

```python
plt.plot(t,x,'c.')
plt.plot(t,x,'b')
plt.xlabel('Temps t(s)')
plt.ylabel('Abscisse x(m)')
plt.title('La trajectoire x(t) avec les valeurs experiementales')
```

```python
m=0.531
sintehta = 0.03/1
z = x * sintehta
g = 9.81
Epp = m*g*z
```

```python
plt.plot(t,Epp, 'r.')
plt.plot(t,Epp,color ='crimson')
plt.xlabel('Temps t(s)')
plt.ylabel('Energie potentielle de pesanteur Epp(J)')
plt.title('La variation de Epp au cours de la trajectoire')
```

```python

v = np.gradient(x,t)
Ec = 0.5 * m* v**2

plt.plot(t,Ec,'.')
plt.plot(t,Ec,'c')
plt.xlabel('Temps t(s)')
plt.ylabel('Energie cintique Ec(J)')
plt.title('La variation de Ec au cours de la trajectoire')
```

```python

plt.plot(t,Epp,'r')
plt.plot(t,Ec, 'g')
plt.plot(t,Epp+Ec,'orange')
plt.xlabel('Temps t(s)')
plt.ylabel('Energies (J)')
plt.title('Les differents energies au cours de la trajectoire')
plt.legend(['Epp','Ec','Em'])
```
